```python
# Collect the abstracts of the top 10000 research papers by using the query "data science"
#installing libraries
from IPython.utils import text
import requests
import json
from typing import Text
import pandas as pd
import csv
#the API request format for this is JSON , hence using the following format
page = ('https://www.semanticscholar.org/api/1/search') 
myobj = {"queryString": "data science",
"page": 1,
"pageSize":10,     # the pagesize is given as 10, as we are getting 1000 abstracts in page 1
"sort": "relevance",
"authors": [],
"coAuthors": [],
"venues": [],
"yearFilter": None,
"requireViewablePdf": False,
"fieldsOfStudy": [],
"useFallbackRankerService": False,
"useFallbackSearchCluster": False,
"hydrateWithDdb": True,
"includeTldrs": True,
"performTitleMatch": True,
"includeBadges": True,
"tldrModelVersion": "v2.0.0",
"getQuerySuggestions": False,
"useS2FosFields": True}
x = requests.post(page, json = myobj).text    # request method:post
y = json.loads(x)
print(y)
results = y["results"]

article_abstracts = []  

for result in results :
  pa = result["paperAbstract"]
  article_abstracts.append(pa["text"])
#text to dataframe 

df = pd.DataFrame({"article_abstracts":article_abstracts})




```

# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python

# creating other column for cleaned abstracts
df['cleaned_abstracts'] = df.loc[:, 'article_abstracts'] 

#Remove noise, such as special characters and punctuations
df['cleaned_abstracts'] = df['cleaned_abstracts'].str.replace('[^\w\s]','') # punctuation
df['cleaned_abstracts']= [x.strip("@#$%?") for x in df['cleaned_abstracts']] # special characters removing 

#to remove the numbers
df['cleaned_abstracts'] = df['cleaned_abstracts'].str.replace('\d+', '') 

#Remove stopwords by using the stopwords list.
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['cleaned_abstracts'] = df['cleaned_abstracts'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

#Lowercase all texts
df['cleaned_abstracts'] = df['cleaned_abstracts'].str.lower()

# Stemming.
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['cleaned_abstracts'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

#Lemmatization.
from textblob import Word
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('punkt') 
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
for i in range(len(df)):
    word_list = nltk.word_tokenize(df.loc[i, "cleaned_abstracts"])
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in word_list])
    df.loc[i, "cleaned_abstracts"] = lemmatized_output
from google.colab import drive
drive.mount('/content/drive')
df.to_csv("//content//drive//MyDrive//article_data.csv")
data = pd.read_csv("//content//drive//MyDrive//article_data.csv")
print(data)
```
